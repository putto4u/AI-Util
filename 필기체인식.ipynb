```python
# 이번에는 MNIST 데이터를 이용해서
# 다중 분류문제를 DNN으로 학습해 보아요!
# 1. 필요한 module import
import numpy as np
import pandas as pd

# 데이터 전처리를 해야해요!
# 1. 결측치
# 2. 이상치
# 3. 정규화
# 이 세가지는 필수적인 처리
from sklearn.preprocessing import MinMaxScaler

# Model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.optimizers import Adam
```

```python
# Raw Data Loading + 전처리
df = pd.read_csv('./train.csv')

# 데이터 분할
# x_data, y_data가 필요해요!
x_data = df.drop('label', axis=1, inplace=False).values
y_data = df['label'].values.reshape(-1,1)   # -1 : 이미지를 1차원으로 하고 행 개수는 직접 자동으로 만들어라
# 일단 이렇게 x_data와 y_data를 분리해서 만들었어요!

# 독립변수 -> 정규화
# 종속변수 -> one-hot encoding을 해야 해요!
#             아..그런데 이 작업이 좀 귀찮아요!
#             Tensorflow는 이 작업을 자기가 알아서 처리!
scaler = MinMaxScaler()
scaler.fit(x_data)
x_data_norm = scaler.transform(x_data)
```

```python
# Model 생성
model = Sequential()
model.add(Input(shape=(784,)))

model.add(Dense(units=32,
                activation='relu'))
model.add(Dense(units=64,
                activation='relu'))

model.add(Dense(units=10,
                activation='softmax'))

model.compile(optimizer=Adam(learning_rate=1e-3),
              loss='sparse_categorical_crossentropy')
model.summary()
```

```python
# 데이터 준비 끝, 모델 구현 끝
# 학습 진행
model.fit(x_data_norm, y_data, epochs=100, verbose=0)
```

```python


# 정확도 확인(생략)
# 예측 실행
# train.csv 에서 일부하나만 가져오기
my_data = x_data_norm[0]  # 첫번째 이미지 데이타 파

result = model.predict(my_data.reshape(1,-1))
print(result)

# [[2.29532775e-22 1.00000000e+00 4.08939022e-26 4.02018180e-37     => 0확률, 1확률 ..... =>1 확율이 제일 높음
#  3.43145249e-28 1.57767655e-27 1.79331342e-29 2.77155956e-28
#  9.46403027e-21 4.52092376e-38]]
```
